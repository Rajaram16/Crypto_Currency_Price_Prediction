```python
#splitting the DATE column
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Original_date'] = df['Date'].dt.day
```

```python
#splitting the DATE column into hour min and sec
df['Hour'] = df['Date'].dt.hour
df['Minute'] = df['Date'].dt.minute
df['Second'] = df['Date'].dt.second
```

```python
df.info()
```

```python
#Change the data type to int for the algo.
df["Low"]=df["Low"].astype(int)
df["Open"]=df["Open"].astype(int)
df["Close"]=df["Close"].astype(int)
df["Volume"]=df["Volume"].astype(int)
df["Marketcap"]=df["Marketcap"].astype(int)
df['Name'] = pd.to_numeric(df['Name'], errors='coerce').fillna(0).astype(int)
df['Symbol'] = pd.to_numeric(df['Symbol'], errors='coerce').fillna(0).astype(int) 
```

```python
X = df[['Year', 'Month', 'Original_date', 'Hour', 'Minute', 'Volume', 'Marketcap','High','Low']]
y = df['Close'] 
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=102,test_size=0.2)
```

```python
import xgboost as xgb
```

```python
RR=xgb.XGBRFRegressor()
```

```python
RR.fit(X_train,y_train)
```

```python
#Training accuracy
RR.score(X_train,y_train)
```

```python
#Testing accuracy
RR.score(X_test,y_test)
```

```python
predictions = RR.predict(X_test)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, predictions, color='blue', alpha=0.6)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='black', linestyle='--')
plt.legend(['Actual','Prediction'])
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Scatter Plot')
plt.show()
```
